```python
import matplotlib.pyplot as plt
```

```python
vg_data = pd.read_csv("vgsales.csv")
vg_data
```

```python
platforms = vg_data['Platform']
subframe = vg_data[['NA_Sales','EU_Sales','Other_Sales']]
```

```python
platforms.value_counts(sort=False).plot.bar()
plt.show()
```

```python
platforms.value_counts(sort=True).plot.hist(orientation = 'vertical')
plt.show()
```

```python
subframe.plot.area(stacked=False)
plt.show()
```

```python
subframe.plot.area(stacked=True)
plt.show()
```

```python
platforms = vg_data['Platform'].head(300)
platforms.value_counts(sort=False).plot.pie(figsize=(8,8))
plt.show()
```

```python
year = vg_data['Year'].head(200)
year.value_counts().plot.kde()
plt.show()
```

```python
print(year)
```

```python
# --
from pandas.plotting import scatter_matrix

scatter_matrix(vg_data,alpha=0.2,figsize=(6,6),diagonal='kde')
plt.show()
```

```python
from pandas.plotting import bootstrap_plot
bootstrap_plot(year,size=50,samples=500,color='blue')
plt.show()
```
